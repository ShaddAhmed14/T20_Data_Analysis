check names manually and see if any further special characters.

```python
import pandas as pd
import json
```

Lets start with our first json file: Match Summary.

```python
with open('t20_json_files/t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

print(data)
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

Score Card is unique to each row, so we can make it our primary key for future table joins.

```python
df_match.rename({'scorecard':'match_id'}, axis=1, inplace=True)
df_match.head()
```

Our 1st CSV is now ready. We can export it.

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv', index = False)
```

Other tables have column containing match data such as Team A vs Team B. We can use match_id as values to these teams as keys so we can link up to other tables.

We must add 2 keys per match_id since team name can be written in either sequence. 

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

print(match_ids_dict)
```

Our next json is Batting Summary.

```python
with open('t20_json_files/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary']) # this appends the lists together.
  
print(all_records)
```

```python
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

Creating out/not_out for easier analysis.

If there is data in dismissal column we regard it as out. Empty value means not out.

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head(11)
```

We no longer require dismissal column.

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10)
```

This is where we match_id_dict to add match_id column to our batting summary dataframe.

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

I noticed some special characters in batsman name. Since there are a just a few of them I will hardcode them out.

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('(c)', ''))
df_batting.head(15)
```

This df is also ready to be exported.

```python
df_batting.to_csv('t20_csv_files/fact_bating_summary.csv', index = False)
```

Next is Bowling Summary.

```python
with open('t20_json_files/t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

Adding match_id to this dataframe the same way we did to batting summary.

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```

Next is Player information.

```python
with open('t20_json_files/t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)

print(data)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

```python
df_players.to_csv('t20_csv_files/dim_players_no_images.csv', index = False)
```
